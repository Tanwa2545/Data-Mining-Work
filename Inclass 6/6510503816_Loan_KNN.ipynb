# Data Pre-Processing

```python
df.drop(['ID', 'ZIP Code'],axis=1,inplace=True)
dn.drop(['Personal Loan'],axis=1,inplace=True)
```

```python
df = df[df['Experience'] >= 0]
df.describe().T
```

```python
y_df = df[['Personal Loan']]
df.drop(['Personal Loan'],axis=1,inplace=True)
X_df = df
with pd.option_context('display.max_rows', 10): display(X_df)
```

```python
X = X_df.values
y = y_df.values
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
new = dn.values
new = scaler.transform(new)
```

# K-NN at k=3

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
neigh.fit(X_norm, y)
```

```python
y_pred = neigh.predict(new)
print(y_pred)
```

# Finding best k for K-NN

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.6, random_state = 1992)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
scores = []
for k in range(1,10):
  neigh = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
  neigh.fit(X_train, y_train.ravel())
  scores.append([k,neigh.score(X_test,y_test)])
```

```python
scores
```

```python
scores.sort(key=lambda x: x[1], reverse=True)
print(f"Best k={scores[0][0]} with accuracy score:{scores[0][1]}")
```

# Using best k and show Confusion Matrix

```python
k = scores[0][0]
neigh = KNeighborsClassifier(n_neighbors=k,metric='euclidean')
neigh.fit(X_train, y_train.ravel())
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = neigh.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
