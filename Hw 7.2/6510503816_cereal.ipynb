# Data Mining Steps:

## a\) Apply K-Mean algorithm with best K

หา best k จาก elbow plot

```python
df_kmean = df[["calories","protein","fat","sodium","fiber","carbo","sugars","potass","vitamins"]]
```

```python
from sklearn.cluster import KMeans
wcss = []
K_rng = range(1, 10)
for K in K_rng:
    km = KMeans(n_clusters=K, init='k-means++')
    km.fit(df_kmean)
    wcss.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared distance')
plt.plot(K_rng, wcss)
```

จากกราฟในที่นี้ จะเลือกใช้ค่า k = 4

```python
km = KMeans(n_clusters=4,init = 'k-means++')
y_predicted = km.fit_predict(df_kmean)
y_predicted
```

ทำการเพิ่ม column cluster เข้าไปใน dataframe

```python
km.fit(df_kmean)
df['cluster'] = km.labels_
df.groupby('cluster').describe().T
```

```python
km.cluster_centers_
```

## b\) Explain the meaning of each cluster by going their description.

```python
centroids = pd.DataFrame(km.cluster_centers_ , columns=df_kmean.columns)
centroids
```

```python
percentage = centroids.div(centroids.sum(axis=0), axis=1) * 100
percentage
```
