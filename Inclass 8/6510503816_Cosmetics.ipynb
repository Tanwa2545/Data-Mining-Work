# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Importing the dataset

```python
df = pd.read_csv('Cosmetics.csv')
df.sample
df.head(10)
```

# Pre-processing

```python
df = df.astype(bool)
df
```

```python
df.drop('Trans. ', axis=1, inplace=True)
df
```

# Model construction using association rules

## Generate all the frequent itemsets with supports = 20%

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

## What is the total number of frequent itemsets?

```python
len(frequent_itemsets)
```

## Show all tha maximal frequent itemsets

```python
frequent_itemsets['Item Set Size'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
frequent_itemsets.drop(columns='Item Set Size', inplace=True)
```

## Generate all the association rules with minimum support = 20% and minimum confidence = 70%

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

# Model deployment

## Show rule with highest confidence

```python
rules.sort_values(by='confidence', ascending=False)
```

### Explain the meaning of this rule

A => B with confidence = C and support = D means if customers buy A, they have a chance of C% to buy B too. But customer only buy both A and B at D%

---   
Example.  
(Mascara) =>	(Eye shadow) wtih confidence = 0.899160 and support = 0.321	means if customer buy Mascara, they have a chance of 89.91% to buy Eye shadow too. But customer only buy both Mascara and Eye shadow at 32.1%

### Explain how support, confidence and lift are calculated?

Support = Freq(A, B) / N  
Confidence =  Freq(A, B) / Freq(A)
