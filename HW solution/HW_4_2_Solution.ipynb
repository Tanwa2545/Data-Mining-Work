# 3.Processing categorical and/or numerical data

## Set roles X and y for supervised-learning

```python
y_df = df['income']
df.drop(['income'],axis=1,inplace=True)
X_df = df
with pd.option_context('display.max_rows', 10): display(X_df)
X = X_df.values
y = y_df.values
```

```python
print(y)
```

## Encoding the Independent Variables

-Categorical to Numerical: Use One-Hot Encoding or Label Encoding.
-โดยสำหรับการ Encoding the Independent Variables เป็นการเปลี่ยนจากcategorical variables ที่มีหลายประเภทเป็น binary vector


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Identify categorical columns
categorical_feature_indices = [X_df.columns.get_loc(col) for col in categorical_data.columns if col in X_df.columns]
```

```python
# Using one hot encoding for encoding independent variables
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(sparse=False), categorical_feature_indices)
    ],
    remainder='passthrough'
)
```

```python
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Encoding the Dependent Variable

- สำหรับการ Encoding the Dependent Variableจะใช้ label encoding เป็นการแทนชื่อประเภทของตัวแปรที่เป็น Label ให้เป็น unique numeric label

```python
# Use label encoder for encoding dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

# 4.Feature Scaling on Training Set & Test Set

-Scaling Numerical Features
ใช้หลักการ Standardization หรือ Normalization ที่เป็น z-score ซึ่งเป็นการใช้การนำข้อมูลมาลบด้วยค่าเฉลี่ยและหารด้วยค่าเบี่ยงเบนมาตราฐาน


```python
numerical_feature_indices = [X_df.columns.get_loc(col) for col in X_df.columns if col not in categorical_data.columns]
```

```python
# Using Standardzation or Normalization for feature scaling numerical data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, numerical_feature_indices] = sc.fit_transform(X_train[:, numerical_feature_indices])
X_test[:, numerical_feature_indices] = sc.transform(X_test[:, numerical_feature_indices])
```

```python
print(X_train)
```

```python
print(X_test)
```

# Train Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
# Train Logistic Regression Model
model = LogisticRegression()
model.fit(X_train, y_train)
# ประเมินประสิทธิภาพของ Model ที่ Test Data
print('Test set accuracy:', model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report, accuracy_score
# Evaluate the model on the test data
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print('Test set accuracy:', accuracy)
print(f"Classification Report:\n{report}")


# Display shapes of the datasets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```
